```python
import matplotlib
#matplotlib.use("Agg")

# import the necessary packages
from tensorflow.keras.preprocessing.image import ImageDataGenerator

from tensorflow.keras.applications.resnet import ResNet50

from tensorflow.keras.layers import Input
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import SGD
from tensorflow.keras.layers import AveragePooling2D, Dropout, Flatten, Dense
#from sklearn.preprocessing import LabelBinarizer
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
#from imutils import paths
import matplotlib.pyplot as plt
import numpy as np
import argparse
import pickle
import cv2
import os
from sklearn.metrics import confusion_matrix
from sklearn.utils.multiclass import unique_labels
```

```python
from sklearn import preprocessing
import numpy as np
import matplotlib.pyplot as plt
import cv2
from sklearn.preprocessing import LabelBinarizer
```

```python
from skimage.io import imread, imshow
from skimage.transform import resize
from skimage.feature import hog
from skimage import exposure
```

```python
import os
dir="Cyclone_Wildfire_Flood_Earthquake_Database/Cyclone_Wildfire_Flood_Earthquake_Database"
classes=[clas.name for clas in os.scandir(dir) if clas.is_dir()]

x,y=[],[]
for clas in classes:
    path=os.path.join(dir,clas)
    for img in os.listdir(path):
        k=cv2.imread(os.path.join(path,img))
        k=cv2.cvtColor(k, cv2.COLOR_BGR2RGB)
        k=cv2.resize(k, (224, 224))
        x.append(k)
        y.append(str(clas))
```

```python
x=np.array(x)
y=np.array(y)
```

```python
imshow(x[0])
```

```python
lb = LabelBinarizer()
y= lb.fit_transform(y)
print(x.shape, y.shape)
```

```python
(trainX, testX, trainY, testY) = train_test_split(x, y,
test_size=0.20, stratify=y, random_state=42)
```

```python
trainAug = ImageDataGenerator(
	rotation_range=30,
	zoom_range=0.15,
	width_shift_range=0.2,
	height_shift_range=0.2,
	shear_range=0.15,
	horizontal_flip=True,
	fill_mode="nearest")
valAug = ImageDataGenerator()
mean = np.array([123.68, 116.779, 103.939], dtype="float32")
trainAug.mean = mean
valAug.mean = mean
```

```python
baseModel = ResNet50(weights="imagenet", include_top=False,
	input_tensor=Input(shape=(224, 224, 3)))

for layer in baseModel.layers:
	layer.trainable = False


headModel = baseModel.output
headModel = AveragePooling2D(pool_size=(7, 7))(headModel)
headModel = Flatten(name="flatten")(headModel)
headModel = Dense(512, activation="relu")(headModel)
headModel = Dropout(0.5)(headModel)
headModel = Dense(len(lb.classes_), activation="softmax")(headModel)

model = Model(inputs=baseModel.input, outputs=headModel)
```

```python
print("[INFO] compiling model...")
opt = SGD(lr=1e-4, momentum=0.9, decay=1e-4 / 25)
model.compile(loss="categorical_crossentropy", optimizer=opt,
	metrics=["accuracy"])
print("[INFO] training head...")
H = model.fit_generator(
	trainAug.flow(trainX, trainY, batch_size=32),
	steps_per_epoch=len(trainX) // 32,
	validation_data=valAug.flow(testX, testY),
	validation_steps=len(testX) // 32,
	epochs=25)
```

```python
from tensorflow.keras.models import load_model
model.save('disaster_classification_pkl.pkl')
```

```python
score_training=model.evaluate_generator((testX,trainY), 
                         steps=10, 
                         verbose=0)
print("Loss of trained model :",score_training[0])
print("Accuracy of trained model :",score_training[1])
```
